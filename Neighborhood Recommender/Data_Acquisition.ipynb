```python
#set up empty lists to collect data
rev_date = []
rev_text = []
rev_rating = []
rev_title = []
```

```python
chromedriver = "/Applications/chromedriver" 
os.environ["webdriver.chrome.driver"] = chromedriver

driver = webdriver.Chrome(chromedriver)
# driver.get(url)
```

```python
def get_rev_data2(url):
    next_url=True
    count=0
    while next_url==True and count<10:
        print(url)
        print(len(rev_date))
        expanded = False
        #get soup, selenium driver
        response = requests.get(url)
        soup = BeautifulSoup(response.text)
        driver.get(url)

        #         rev_rating.append(i.find(class_="reviewSelector")["id"])

        #get ratingDate
        for i in soup.find_all(class_="prw_rup prw_reviews_review_resp"):
            rev_date.append(i.find(class_="ratingDate")["title"])
        #get review rating
            rev_rating.append(i.find(class_="ui_column is-9").findChildren()[0]["class"])

        #get review title
            rev_title.append(i.find(class_="noQuotes").text)

        #get review text    
            time.sleep(2)
            end = i.find(class_="partial_entry").text[-7:]
            
            my_id = i.find(class_="reviewSelector")["id"]
            print(my_id)
            if end=="...More" and expanded==False:

                review = driver.find_element_by_id(my_id)
                time.sleep(3)
                review.find_element_by_css_selector(".taLnk.ulBlueLinks").click()
                time.sleep(2)
                new_review = driver.find_element_by_id(my_id)
                entry = new_review.find_element_by_class_name("partial_entry")

                rev_text.append(entry.text)
                expanded=True
            else:
                new_review = driver.find_element_by_id(my_id)
                entry=new_review.find_element_by_class_name("partial_entry")
                rev_text.append(entry.text)
        a = soup.find(class_="nav next taLnk ui_button primary")["href"]
        if a == "":
            next_url=False
        else:
            url = "https://www.tripadvisor.com"+a
            count+=1
```

```python
get_rev_data2("https://www.tripadvisor.com/Attraction_Review-g60713-d110240-Reviews-or100-Nob_Hill-San_Francisco_California.html")
```
