```python
from yelpapi import YelpAPI
yelp_api = YelpAPI("###apikey")

import pandas as pd
```

```python
data = pd.read_csv("DOHMH_New_York_City_Restaurant_Inspection_Results.csv.crdownload")
```

```python
data.head()
```

```python
data["address"] = data["BUILDING"] +" " + data["STREET"]
```

```python
data.head()
```

```python
data["ZIPCODE"] = data["ZIPCODE"].apply(lambda x: int(x))
```

```python
data = data[data["ZIPCODE"]>0]
```

```python
data.shape
```

```python
query_data = data[["DBA","address", "ZIPCODE"]]
```

```python
query_data.sort_values(by="DBA", inplace=True)
```

```python
query_data.drop_duplicates(inplace=True)
```

```python
query_data.shape
```

```python
query_data.dropna(inplace=True)
```

```python
query_data["id"] = 0
query_data["categories"]=0
query_data["coordinates"]=0
query_data["hours"]=0
query_data["image_url"]=0
query_data["is_closed"]=0
query_data["photos"]=0
query_data["price"]=0
query_data["rating"]=0
query_data["review_count"]=0
query_data["transactions"]=0
```

```python
query_data
```

```python
def get_ids(start, stop):
    for i in range(start, stop):
        rest = query_data.iloc[i,0]
        addr = query_data.iloc[i,1]
        zip_co= query_data.iloc[i,2]
        info = yelp_api.business_match_query(name=rest,city="New York", address1 = addr, state="NY", zip_code =zip_co ,country="US")
        try:
            query_data.iloc[i,3] = info["businesses"][0]["id"]
        except:
            continue
```

```python
def get_info(start, stop):
    for i in range(start, stop):
        id_curr = query_data.iloc[i,3]
        if id_curr == 0:
            continue
        try:
            curr = yelp_api.business_query(id=id_curr)
        except:
            continue
        try:
            query_data.iloc[i,4] = str(curr["categories"])
        except:
            pass
        try:
            query_data.iloc[i,5] = str(curr["coordinates"])
        except:
            pass
        try:
            query_data.iloc[i,6] = str(curr["hours"])
        except:
            pass
        try:
            query_data.iloc[i,7] = str(curr["image_url"])
        except:
            pass
        try:
            query_data.iloc[i,8] = str(curr["is_closed"])
        except:
            pass
        try:
            query_data.iloc[i,9] = str(curr["photos"])
        except:
            pass
        try:
            query_data.iloc[i,10] = str(curr["price"])
        except:
            pass
        try:
            query_data.iloc[i,11] = str(curr['rating'])
        except:
            pass
        try:
            query_data.iloc[i,12] = str(curr['review_count'])
        except:
            pass
        try:
            query_data.iloc[i,13] = str(curr['transactions'])
        except:
            pass
```
